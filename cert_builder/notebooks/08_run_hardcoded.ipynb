# Hardcoded Content Loader
Loads static content from hardcoded_content.yaml.

```python
import sys
from pathlib import Path

ROOT = Path.cwd() if Path.cwd().name == 'certification_framework' else Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
```

```python
import json
from scripts.computation.hardcoded_loader import load_all

NB_DATA = ROOT / 'notebooks' / 'data'
result = load_all()

# Save to notebook-local data
(NB_DATA / 'hardcoded.json').write_text(
    json.dumps(result, indent=2, default=str, ensure_ascii=False), encoding='utf-8'
)

hc = result['hardcoded']
print('=== Hardcoded Content ===')
print(f'\nDefinitions ({len(hc["definitions"])}):')
for k in hc['definitions']:
    print(f'  {k}: {str(hc["definitions"][k])[:70]}')

print(f'\nSection Intros ({len(hc["section_intros"])}):')
for k in hc['section_intros']:
    print(f'  {k}: {str(hc["section_intros"][k])[:70]}')

print(f'\nMethodology Bullets ({len(hc["methodology_bullets"])}):')
for b in hc['methodology_bullets']:
    print(f'  - {b[:70]}')
```
