# Assessment Formatter
Runs the assessment formatter independently.

```python
import sys
from pathlib import Path

ROOT = Path.cwd() if Path.cwd().name == 'certification_framework' else Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
```

```python
import json
from scripts.computation.assessment_formatter import build_from_file

NB_DATA = ROOT / 'notebooks' / 'data'
phase1_path = NB_DATA / 'parsed_context.json'
result = build_from_file(phase1_path)

# Save to notebook-local data
(NB_DATA / 'assessments.json').write_text(
    json.dumps(result, indent=2, default=str, ensure_ascii=False), encoding='utf-8'
)

print(f'=== Assessments ({len(result["assessments"])} categories) ===')
for cat, blocks in result['assessments'].items():
    print(f'\n  {cat}: {len(blocks)} blocks')
    for b in blocks:
        print(f'    [{b.get("severity", "?")}] {b.get("title", b.get("text", ""))[:70]}')
```

```python
raw = json.loads((ROOT / 'data' / 'input' / 'aggregated_scorecard_output.json').read_text(encoding='utf-8'))

print('=== Validation: Assessments vs Input ===')
all_ok = True
cat_count = len(raw['fault_category_scorecards'])

ok = 'PASS' if len(result['assessments']) == cat_count else 'FAIL'
if ok == 'FAIL': all_ok = False
print(f'  {ok} category count: {len(result["assessments"])} (expected {cat_count})')

# Each category should have assessment blocks
for cat, blocks in result['assessments'].items():
    ok = 'PASS' if len(blocks) > 0 else 'FAIL'
    if ok == 'FAIL': all_ok = False
    print(f'  {ok} {cat}: {len(blocks)} blocks (expected > 0)')

print(f'\nResult: {"ALL CHECKS PASSED" if all_ok else "SOME CHECKS FAILED"}')
```
